```python
import qrcode
from PIL import Image
```

<h1> QR Generator </h1>

Qr without logo

```python
#function
def qr(link,fill_color,back_color):
    qr= qrcode.QRCode(version = 1, box_size = 8, border = 4)
    qr.add_data(link)
    qr.make()
    qr=qr.make_image(fill_color = fill_color, back_color = back_color)
    return qr
```

```python
qr("https://www.youtube.com/watch?v=EqMwD8HBjEE","white","pink")
```

Qr with logo

```python
#Logo
logo= Image.open("sakura.jpg")
```

```python
#function
def qrlogo(logo,logo_size,link,fill_color,back_color):
    # adjust image size
    wpercent = (logo_size/float(logo.size[0]))
    hsize = int((float(logo.size[1])*float(wpercent)))
    logo = logo.resize((logo_size, hsize), Image.ANTIALIAS)
    qrlogo = qrcode.QRCode(error_correction=qrcode.constants.ERROR_CORRECT_H)
    # adding URL to qrlogo
    qrlogo.add_data(link)
    # generating qrlogo
    qrlogo.make()
    # adding color to qrlogo
    qrlogo_img=qrlogo.make_image(fill_color = fill_color, back_color = back_color).convert('RGB')
    # position of the logo to center of qrlogo
    logo_x_position = (qrlogo_img.size[0] - logo.size[0]) // 2
    logo_y_position = (qrlogo_img.size[1] - logo.size[1]) // 2
    logo_position = (logo_x_position, logo_y_position)
    # inserting logo image into qrlogo image
    qrlogo_img.paste(logo, logo_position)
    # saving qrlogo image
    qrlogo_img.save('logoGenerated.png')
    return qrlogo_img
```

```python
qrlogo(logo,150,"https://www.youtube.com/watch?v=EqMwD8HBjEE","red","white")
```
